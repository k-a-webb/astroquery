```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
from astropy.io import ascii
```

```python
table = pd.read_excel('CometEls.xlsx')
#table
```

```python
ecc = table['eccentricity']
q = table['perihelion']
```

```python
a = table['perihelion']/(1 - table['eccentricity'])  # a=q/(1-e)
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    plt.scatter(a, ecc)
    plt.ylim(0,1.2)
    plt.ylabel('eccentricity')
    plt.xlim(0, 20)

    plt.subplot(212)
    plt.scatter(a, q)
    plt.ylim(0,14)
    plt.ylabel('perihelion')
    plt.xlim(0, 500)
    plt.xlabel('semimajor axis (AU)')
```
