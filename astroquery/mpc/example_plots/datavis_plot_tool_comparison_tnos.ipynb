```python
import seaborn as sns

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import bokeh.plotting as bp
from bokeh.objects import Range1d

from astroquery import mpc

%matplotlib inline
sns.set_context('poster')  # I like big plots and I cannot lie
```

Let's get some data from the MPC: everything with a > 15.
Note that the data come back as a list of dicts, which is a valid input format for a pandas DataFrame, and allows the data to be dealt with as "tidy data" (each row is a TNO, each column an orbital property of that TNO).

```python
mpccenter = mpc.Mpc()
mpccenter._login('mpc_ws', 'mpc!!ws')
tnos = pd.DataFrame(mpccenter.query_parameters('semimajor_axis_min 20',
                                  return_request='semimajor_axis, inclination, eccentricity, name, designation'))
a_min = 20
tnos["perihelion"] = tnos['semimajor_axis']*(1. - tnos['eccentricity'])  # q = a(1-e)
print(len(tnos))
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    plt.scatter(tnos["semimajor_axis"], tnos["eccentricity"])
    plt.ylim(0,1)
    plt.ylabel('eccentricity')
    plt.xlim(a_min, 100)

    plt.subplot(212)
    plt.scatter(tnos["semimajor_axis"], tnos["inclination"])
    plt.ylim(0,90)
    plt.ylabel('inclination (deg)')
    plt.xlim(a_min, 100)
    plt.xlabel('semimajor axis (AU)')
```

This basic plot is nice and I'm quite happy with it as the default what-you-see of TNOs. 
Nothing about this other than the use of Seaborn defaults is different to anything I'd do normally with matplotlib.
But they're good defaults. Saves the fifteen lines of plot tweaking I normally script!

Now let's see if perihelion can be assigned as a hue value to give a third level of information to this standard aei plot.
First let's see how perihelion is distributed, no surprises there. And what colourmap I want to assign.

```python
peribins = [0, 25, 40, 50, 100]
tnos.hist(column="perihelion", bins=peribins)
d = np.digitize(tnos["perihelion"], peribins) 
tnos['peribin'] = [str(peribins[d[n]-1])+'-'+str(peribins[d[n]]) for n in range(len(tnos))]  # nice text axis labels
```

Hey look Pandas has a built-in histogram function. Still need to use numpy to get the values in each bin, but it's one line of code.

```python
tnos_subset = tnos[800:850]  # try with a subsample because using all of them crashes the kernel
g = sns.FacetGrid(tnos_subset, hue="perihelion", margin_titles=True)
g.map(plt.scatter, "semimajor_axis", "eccentricity", alpha=.7, cmap=sns.cubehelix_palette(8, start=.5, rot=-.75))
plt.xlim(a_min, 100)
plt.ylim(0,1)
```

So this doesn't work particularly well. Specifically, I can't seem to find a way to map the histogram bins across to Hue bins within the way seaborn is set up. And the axis crushing done by FacetGrid's defaults is a pain; this probably isn't what it's meant for. Let's try sns.PairGrid, which also takes a hue argument.
http://web.stanford.edu/~mwaskom/software/seaborn/generated/seaborn.PairGrid.html#seaborn.PairGrid

```python
pg = sns.PairGrid(tnos, x_vars=["semimajor_axis"], y_vars=["eccentricity", "inclination"], 
                  aspect=4, hue="peribin", palette="coolwarm")
plt.xlim(a_min, 100)
pg.axes[0][0].set_ylim([0,1.])
#gg.axes[0][1].set_ylim([0,90.])  # Calling this instead, which would be normal, gives an index out of range error
plt.ylim(0,90)  # but this fixes it just fine on the second plot. Huh.
pg.map(plt.scatter)       #, c=tnos["perihelion"], cmap="coolwarm")  # wish this worked, but get odd errors
pg.add_legend(title='perihelion (AU)')  # has to be after the map, but breaks if the legend is passed continuous data rather than categorical.
pg.savefig('/Users/michele/Desktop/tnos_coolwarm.pdf')
```

This is a lot better. Nice easy subplotting. Does take longer to draw than the original plt.scatter() if continuous data are put through the hue= argument, but is just as fast if a discrete (binned) variable is plotted instead.
